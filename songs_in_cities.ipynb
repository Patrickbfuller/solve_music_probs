```python
from selenium.webdriver import Chrome
import scrape
import pandas as pd
%load_ext autoreload
%autoreload 2
```

```python
browser = Chrome()
url = 'http://everynoise.com/everyplace.cgi?vector=city&scope=North%20America'
```

```python
browser.get(url)
```

```python
cities = browser.find_elements_by_xpath('/html/body/table/tbody/tr/td[1]/table/tbody/tr')
```

```python
for city in cities:
    for td in city.find_elements_by_css_selector('td'):
        print(td.text)
    link = city.find_element_by_css_selector('td.note > a')
    print('\t', link.get_attribute('href'))
```

```python
cities_list = []
for city in cities:
    new_row = []
    for td in city.find_elements_by_css_selector('td'):
        new_row.append(td.text)
    link = city.find_element_by_css_selector('td.note > a')
    new_row.append(link)
    cities_list.append(new_row)
```
